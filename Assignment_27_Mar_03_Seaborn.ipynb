### Q1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. **Box Plot** : 
> A box plot is a graphical representation of data that shows the distribution of a dataset by indicating the median, quartiles, and outliers. It is used to identify the spread and skewness of the data.
2. **Heatmap** : 
> A heatmap is a graphical representation of data that uses color-coding to represent the values of a two-dimensional dataset. It is used to identify patterns in the data, such as correlations or clusters.
3. **Pair Plot** : 
> A pair plot is a graphical representation of data that shows pairwise relationships between different variables in a dataset. It is used to identify correlations between variables and to explore the relationships between them.
4. **Violin Plot** : 
> A violin plot is a graphical representation of data that combines elements of a box plot and a kernel density plot. It is used to visualize the distribution of a dataset and show its density.
5. **Joint Plot** : 
> A joint plot is a graphical representation of data that shows the relationship between two variables using a scatter plot and a histogram. It is used to explore the relationship between two variables and to identify correlations between them.

These are just a few examples of the types of plots that can be created using Seaborn. Seaborn offers many other visualizations that can be used to explore and analyze different types of data.

### Q2. Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. 

**Note**:  timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
fmri
```

```python
fmri['event'].value_counts()
```

```python
fmri['region'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
sns.lineplot(x = "timepoint", y = "signal", data = fmri, style = "event", hue = "region")
plt.title('Signal vs Time for Different Events and Regions')
plt.show()
```

### Q3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

**Note**:  pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic
```

```python
import matplotlib.pyplot as plt

# Creating a figure with 1 rows and 2 columns
fig, axes = plt.subplots(1,2, figsize=(10, 5))

# create chart in each subplot
sns.boxplot(data = titanic, x = "pclass", y = "age", ax = axes[0])
axes[0].set_title("Passenger Class Vs Passanger Age")

sns.boxplot(data = titanic, x = "pclass", y = "fare", ax = axes[1])
axes[1].set_title("Passenger Class Vs Fare")
```

### Q4.  Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset. 

```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
diamonds
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
sns.histplot(data = diamonds, x = "price", hue = "cut")
plt.title("Price distribution for various diamond cuts")
plt.show()
```

### Q6. Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
flights = sns.load_dataset("flights")
flights
```

```python
# Pivot the dataset to create a pivot table with month on the x-axis, year on the y-axis, and passengers as values
flights_matrix = flights.pivot_table(index="year", columns="month", values="passengers")
flights_matrix
```

```python
import matplotlib.pyplot as plt
sns.heatmap(data = flights_matrix, 
            annot = True, 
            fmt = "d",
            linewidths = 0.5,
            linecolor = "white")
plt.title("Passengers Per Month from year 1949 to 1960")
plt.show()
```
