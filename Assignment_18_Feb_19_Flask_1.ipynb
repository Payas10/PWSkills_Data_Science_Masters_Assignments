### 1. What is Flask Framework? What are the advantages of Flask Framework?

### 2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
#import the Flask library
from flask import Flask

#Create the Flask instance and pass the Flask constructor the path of the currect module
app = Flask(__name__)

#Default route added using a decorator for view function
@app.route('/')  
def hello():  
    return 'Hello World!!'  

#Start with flask web app with debug as True only if this is the starting page
if __name__ == '__main__' :  
    app.run(host = "0.0.0.0", debug = True)
```

### 3. What is App routing in Flask? Why do we use app routes?

#### WHAT?
* App Routing means mapping the URLs to a specific function that will handle the logic for that URL. Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler. 
* **Example** : In our application, the URL (“/”) is associated with the root URL. So if our site’s domain was www.example.org and we want to add routing to “www.example.org/hello”, we would use “/hello”. 
* To bind a function to an URL path we use the *app.route decorator*.


#### WHY?
* Routing is the mechanism which helps in mapping the URL directly to the code that creates the webpage and bind an URL to a function.
* It helps in better management of the structure of the webpage and increases the performance of the site considerably and further enhancements or modifications become really straight forward.

### 4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:

    Company Name: ABC Corporation

    Location: India

    Contact Detail: 999-999-9999

```python
#import the Flask library
from flask import Flask

#Create the Flask instance and pass the Flask constructor the path of the currect module
app = Flask(__name__)

#Default route added using a decorator for view function
@app.route('/welcome')  
def welcome():  
    return 'Welcome to ABC Corporation'  
  
@app.route('/')
def details():
    return '''Company Name: ABC Corporation </br>
              Location: India </br>
              Contact Detail: 999-999-9999'''


#Start with flask web app with debug as True only if this is the starting page
if __name__ == '__main__' :  
    app.run(host = "0.0.0.0", debug = True)
```

### 5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

* The url_for() method, is used in Flask to prepare a URL (URL building), for a function dynamically, such that, changing URLs, in the application, is avoided. 
* It accepts, the name of the view function, as the first argument, and, any number of keywords, to be sent(to the view function), as the second argument. 
* Syntax -
> *url_for( ‘function_name’,  key = value )*

```python
# python code to demonstrate the working of the url_for() function

#import the Flask library
from flask import Flask, redirect, url_for, request

#Create the Flask instance and pass the Flask constructor the path of the currect module
app = Flask(__name__)

#Default route added using a decorator for view function
@app.route('/dsm')  
def dsm():  
    return 'Data Science Masters'  
  
@app.route('/webdev')  
def webdev():  
    return 'Web Development'  
  
@app.route('/javadsa')  
def javadsa():  
    return 'Java with DSA'

@app.route('/')
def no_such_course_avail():
    return 'No such course is available'

@app.route('/course/<name>')  
def course(name):  
    if name == 'dsm':  
        return redirect(url_for('dsm'))  
    if name == 'webdev':  
        return redirect(url_for('webdev'))  
    if name == 'javadsa':  
        return redirect(url_for('javadsa'))
    else:
        return redirect(url_for('no_such_course_avail'))

#Start with flask web app with debug as True only if this is the starting page
if __name__ == '__main__' :  
    app.run(host = "0.0.0.0", debug = True)
```
