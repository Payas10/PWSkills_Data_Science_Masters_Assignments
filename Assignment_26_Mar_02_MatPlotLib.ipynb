### Q2. What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
#### Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

#### WHAT?
* Scatter plots are used to observe relationship between variables and uses dots to represent the relationship between them. 
* Scatter plots are widely used to represent relation among variables and how change in one affects the other.
* The scatter() method in the matplotlib library is used to draw a scatter plot. 

```python
# plotting scatter plot using generated data

import matplotlib.pyplot as plt

plt.figure(figsize=(6,3))
plt.scatter(x,y)
plt.xlabel("variable 'X' values are on x-axis")
plt.ylabel("variable 'Y' values are on y-axis")
plt.title("Q2 - X vs Y scatter plot")
plt.grid()
```

### Q3.  Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data.

```python
import numpy as np

# for line 1: 
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# for line 2:
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# for line 3: 
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

# for line 4: 
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

#### WHY?
* subplot() function adds subplot to a current figure at the specified grid position.
* The subplot() function in Matplotlib is used to create multiple plots in a single figure. By sepecifing the no. of rows and columns.
* It is similar to the subplots() function however unlike subplots() it adds one subplot at a time. So to create multiple plots you will need several lines of code with the subplot() function.
* Another drawback of the subplot function is that it deletes the preexisting plot on your figure.

```python
# defining figure size
plt.figure(figsize=(8, 8))

# Creating the first subplot
plt.subplot(2,2,1)
plt.plot(x1, y1)
plt.title("Line 1")

# Creating the second subplot
plt.subplot(2,2,2)
plt.plot(x2, y2)
plt.title("Line 2")

# Creating the first subplot
plt.subplot(2,2,3)
plt.plot(x3, y3)
plt.title("Line 3")

# Creating the first subplot
plt.subplot(2,2,4)
plt.plot(x4, y4)
plt.title("Line 1")

# Displaying the plot
plt.show()
```

#### By using *subplots()*

```python
# By using plt.subplots()

# Creating a figure with 2 rows and 2 columns
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(8, 8))

# Creating the first subplot
ax1.plot(x1, y1)
ax1.set_title('Line 1')

# Creating the second subplot
ax2.plot(x2, y2)
ax2.set_title('Line 2')

# Creating the third subplot
ax3.plot(x3, y3)
ax3.set_title('Line 3')

# Creating the fourth subplot
ax4.plot(x4, y4)
ax4.set_title('Line 4')

# Displaying the plot
plt.show()
```

### Q4. What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

#### WHAT is BAR Plot?
* A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. 
* The bar plots can be plotted horizontally or vertically. 
* A bar chart describes the comparisons between the discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,3))
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Profit comparision of Companies")
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.barh(company, profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Profit comparision of Companies")
plt.show()
```

### Q5. What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

#### WHAT is BOX Plot?
* A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like **minimum, first quartile, median, third quartile** and **maximum**. 
* In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. 
* Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

```python
plt.boxplot([box1, box2])
plt.xlabel("categories on X-axis")
plt.ylabel("Values on Y-axis")
plt.title("Box Plot")
plt.show()
```
